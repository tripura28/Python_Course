```python
import matplotlib.pyplot as plt
```

```python
x=[2,3,4,5,6,7]
y=[1,2,3,4,5,6]
y1=[9,8,5,6,7,3]
y2=[4,5,6,7,8,9]
y3=[4,3,5,5,6,2]
plt.plot(x,y)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.show()
```

```python
plt.subplot(3,5,2)
plt.suptitle("Data") #super title
plt.plot(x,y1)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graph1")

plt.subplot(2,2,2)
plt.title("graph2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y2)
plt.show()
```

```python
plt.subplot(6,2,2)
plt.suptitle("Data")
plt.title("graph3")
plt.plot(y2,y3)
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.subplot(3,5,2)
plt.plot(x,y)
plt.title("graph4")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
